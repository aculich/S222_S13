```python
import pickle
import re
file = open('data/new_comments.pickle','rb') # the comments dictionary has the comments for all videos in .train and .test
raw_comments = pickle.load(file)
stripped_comments = {}
file.close()
pattern = re.compile(r'[,\.\?\"\!:;\)\(]') # regex to strip punctuation (and replace it with a space)
apostrophe = re.compile(r'\'') # strip this separately because we don't replace it with a space
for id in raw_comments: 
    if raw_comments[id] == None or len(raw_comments[id]) == 0:
        continue # delete the entries which don't have any comments
    stripped_comments[id] = []
    for comment in raw_comments[id]: # strip punctuation, replace with space (except no space for apostrophe)
        stripped_comments[id].append( re.sub(apostrophe,"",re.sub(pattern," ",comment)))
```
