```python
import nltk

stopwords = nltk.corpus.stopwords.words('english')
def nop(word):
    return word

stemming = True
CountWordOnce = True

stm = nltk.stem.porter.PorterStemmer()
if stemming:
    stem = stm.stem
else:
    stem = nop
```

```python
import pickle
import re
file = open('Data/new_comments.pickle','rb') # the comments dictionary has the comments for all videos in .train and .test
comments = pickle.load(file)
file.close()
comments0 = {}
pattern = re.compile(r'[,\.\?\"\!:$&-=;\)\(]') # regex to strip punctuation (and replace it with a space)
apostrophe = re.compile(r'\'') # strip this separately because we don't replace it with a space
for id in comments: 
    if comments[id] == None:
        comments0[id]=[]
        continue # delete the entries which don't have any comments
    modcomment=[]
    for comment in comments[id]:
        comment = comment.replace("\xef\xbb\xbf", " ")
        comment = re.sub(apostrophe,"",re.sub(pattern," ",comment)).lower().split()
        comment = [stem(w) for w in comment if w not in stopwords]
        comment = " ".join(comment)
        modcomment.append(comment)
    comments0[id]=modcomment
pd_comments = pd.DataFrame({'comments': comments0})
wlc_clean = pd.merge(won_lost,pd_comments,left_index=True,right_index=True) # now merge the comments in 
```

```python
# Function to count the number of comments containing a word for a given id.

def M(id,w):
    count=1
    commentlist=wlc_clean['comments'][id]
    if commentlist:
        n=len(commentlist)
        for i in range(0,n):
            comment=commentlist[i].split() 
            if w in comment:
                count += 1
    return(count)
```

```python
def p(i,w): # i=0,1
    m=0
    if i == 0:
        for id in lose_ids:
            m += M(id,w)
        prob=float(m)/float(n0)
    elif i == 1:
        for id in win_ids:
            m += M(id,w)
        prob=float(m)/float(n1)
    return(prob)
```

```python
import pandas as pd
testing_pairs = pd.read_csv('Data/comedy_comparisons.test', names=['left', 'right', 'winner'])
```

```python
import math
def logBayes(id):
    data = wlc_clean['comments'][id]
    n = len(data)
    terms = set()
    for row in data:
        terms = terms | set(row.split())
    posterior = 0.0
    for w in terms:
        m=M(id,w)
        p1=p(1,w)
        p0=p(0,1)
        logscore = m*math.log(p1/p0)+(n-m)*math.log((1-p1)/(1-p0))
        posterior+=logscore
    return(posterior)
```

```python
# Score the model
n=len(testing_pairs)
correct=0
for i in range(0,n):
    if (logBayes(testing_pairs['left'][i])>logBayes(testing_pairs['right'][i]) & (testing_pairs['winner'][i]=='left'):
        correct +=1
```
