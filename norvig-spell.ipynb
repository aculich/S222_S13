```python
# based on the Norvig Spelling Corrector: http://norvig.com/spell-correct.html

# CONFIGURATION SECTION

# You can adjust this to determine the total number of videos to pull
# comments from. You probably want to keep this to less than 500 or it
# will start taking a LONG time! The default value of 20 is quick
# enough and will give you the flavor of how this will run with the
# default training set from Norvig, which is big.txt
video_num = 20
training_set  = 'Data/big.txt'
pickled_comments = 'Data/new_comments.pickle'
```

```python
import re, collections
from collections import Counter
import pickle
import logging
from   logging   import debug, info, error
import sys, os, os.path
import getopt
from   itertools import islice
import pdb
import pandas as pd
```

```python
def flatten(l):
    return [item for sublist in l for item in sublist]
def take(n, iterable):
    "Return first n items of the iterable as a list"
    return list(islice(iterable, n))

def words(text): return re.findall("[a-z]+", text.lower())

def train():
    return Counter(words(file(training_set).read()))
```

```python
# Make sure we uncompress into the Data directory if we haven't
# already before we train and open comment data
if not (os.path.exists(training_set) and os.path.exists(pickled_comments)):
    info('Uncompressing data')
    os.system('./uncompress_data')
```

```python
NWORDS = train()
video_comments = pickle.load(open(pickled_comments))
NMAXLEN = max(map(len, NWORDS))
alphabet = 'abcdefghijklmnopqrstuvwxyz'
```

```python
def edits1(word):
   splits     = [(word[:i], word[i:]) for i in range(len(word) + 1)]
   deletes    = [a + b[1:] for a, b in splits if b]
   transposes = [a + b[1] + b[0] + b[2:] for a, b in splits if len(b)>1]
   replaces   = [a + c + b[1:] for a, b in splits for c in alphabet if b]
   inserts    = [a + c + b     for a, b in splits for c in alphabet]
   return set(deletes + transposes + replaces + inserts)

def known_edits2(word):
    return set(e2 for e1 in edits1(word) for e2 in edits1(e1) if e2 in NWORDS)

def known(words): return set(w for w in words if w in NWORDS)

def correct(word):
    # just return the word without correction if it is 2 chars longer
    # than any word in our dictionary, otherwise it will get stuck
    # doing very long calculations unnecessarily.
    if len(word) > NMAXLEN+2:
       return word
    candidates = known([word]) or known(edits1(word)) or known_edits2(word) or [word]
    return max(candidates, key=NWORDS.get)
```
